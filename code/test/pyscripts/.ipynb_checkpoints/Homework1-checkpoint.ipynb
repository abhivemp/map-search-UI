```python
from datetime import datetime
def year_from_date(date):
    return datetime.strptime(date, "%Y-%m-%dT%H:%M:%SZ").year
```

```python
years = { }
for obj in data['results']:
    year = year_from_date(obj['releaseDate'])
    if year in years.keys():
       years[year]+=1
    else:
        years[year] = 1
        
print(years)
```
